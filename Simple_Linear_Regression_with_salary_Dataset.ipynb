```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("Salary_dataset.csv")
```

```python
data.info()
```

```python
sns.scatterplot(x=data["YearsExperience"],y=data["Salary"],data=data)
plt.title("Salary vs Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
# Split The Independent And Dependent Variable

X = pd.DataFrame(data.iloc[ : , 1])
y = pd.DataFrame(data.iloc[ : , -1])
```

```python
# Split The Trining and testing Data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Make the model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model
```

```python
model.fit(X_train,y_train)
```

```python
# Predict The y for X_test
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print(f"R2_Score : {r2_score(y_test,y_pred)}")
print(f"Mean Absolute Error : {mean_absolute_error(y_test,y_pred)}")
print(f"Mean Squared Error : {mean_squared_error(y_test,y_pred)}")
```

```python
plt.scatter(X_test,y_test,color="red",label="Original Data Point")
plt.scatter(X_test,y_pred,color="blue",label="Predicted Data Point")
plt.plot(X_test,y_pred,color="pink",label="Regression Line")
plt.title("Salary vs Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
